```python
import plotly
import pandas as pd

import plotly.plotly as py
import numpy as np
```

```python
data = pd.read_csv('vg_sales_rating.csv')
data = data[(~data.Name.isnull()) & (data.Year_of_Release <= 2016)
            & (~data.Genre.isnull()) & (~data.Publisher.isnull())]

genre_included = ['Action', 'Sports', 'Shooter', 'Role-Playing', 'Platform', 'Misc', 'Racing']
data = data[data.Genre.isin(genre_included)]

data = data.groupby(['Year_of_Release', 'Genre']).sum().loc[:,['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
data.columns = ['Sales_NA', 'Sales_EU', 'Sales_JP', 'Sales_Other']
data = data.reset_index(level=[0,1])
data['id'] = data.index
data.head(5)
```

```python
melt_data = pd.melt(data,
                    id_vars=['Year_of_Release', 'Genre'],
                    value_vars=['Sales_NA', 'Sales_EU', 'Sales_JP', 'Sales_Other'],
                    var_name='GeoLocation', value_name='Sales')
melt_data.head(), melt_data.shape
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_us_cities.csv')
df.head()

df['text'] = df['name'] + '<br>Population ' + (df['pop']/1e6).astype(str)+' million'
limits = [(0,2),(3,10),(11,20),(21,50),(50,3000)]
colors = ["rgb(0,116,217)","rgb(255,65,54)","rgb(133,20,75)","rgb(255,133,27)","lightgrey"]
cities = []
scale = 5000

class MyDict(dict):
    def __init__(self, *args, **kw):
        super(MyDict,self).__init__(*args, **kw)
        self.__setitem__('v1', False)
        self.__setitem__('v2', False)
        
    def __getitem__(self, key):
        if key == 'visible':
            return self.__getitem__('v1') and self.__getitem__('v2')
        else:
            return super(MyDict,self).__getitem__(key)

for i in range(len(limits)):
    lim = limits[i]
    df_sub = df[lim[0]:lim[1]]
    city = MyDict(
        type = 'scattergeo',
        locationmode = 'USA-states',
        visible=False,
        lon = df_sub['lon'],
        lat = df_sub['lat'],
        text = df_sub['text'],
        marker = dict(
            size = df_sub['pop']/scale,
            color = colors[i],
            line = dict(width=0.5, color='rgb(40,40,40)'),
            sizemode = 'area'
        ),
        name = '{0} - {1}'.format(lim[0],lim[1]) )
    cities.append(city)
    
cities[2]['v1'] = True
cities[3]['v2'] = True


steps1 = []
for i in range(len(cities)):
    step = dict(
        method = 'restyle',
        args = ['v1', [False] * len(cities)],
    )
    step['args'][1][i] = True # Toggle i'th trace to "visible"
    steps1.append(step)
    
steps2 = []
for i in range(len(cities)):
    step = dict(
        method = 'restyle',
        args = ['v2', [False] * len(cities)],
    )
    step['args'][1][i] = True # Toggle i'th trace to "visible"
    steps2.append(step)

sliders = [dict(
    active = 2,
    currentvalue = {"prefix": "Frequency: "},
    pad = {"t": 0},
    steps = steps1
), dict(
    active = 3,
    currentvalue = {"prefix": "FFR: "},
    pad = {"t": 100},
    steps = steps2
)]

layout = dict(sliders=sliders)

fig = dict(data=cities, layout=layout)

py.iplot(fig, filename='Sine Wave Slider', validate=False)
```
