```python
import plotly
import pandas as pd

import plotly.plotly as py
import numpy as np
```

```python
data = pd.read_csv('vg_sales_rating.csv')
data = data[(~data.Name.isnull()) & (data.Year_of_Release <= 2016)
            & (~data.Genre.isnull()) & (~data.Publisher.isnull())]

data = data.groupby(['Year_of_Release']).sum().loc[:,['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
data.columns = ['Sales_NA', 'Sales_EU', 'Sales_JP', 'Sales_Other']
data = data.reset_index(level=[0])
data['id'] = data.index
data.head(5)
```

```python
melt_data = pd.melt(data,
                    id_vars=['Year_of_Release'],
                    value_vars=['Sales_NA', 'Sales_EU', 'Sales_JP', 'Sales_Other'],
                    var_name='GeoLocation', value_name='Sales')

GeoMap = {
    'Sales_NA': (-100, 45),
    'Sales_EU': (15, 50),
    'Sales_JP': (140, 37),
    'Sales_Other': (50, -25)
}
GPS = melt_data.loc[:, 'GeoLocation'].map(GeoMap)
melt_data['Lon'] = GPS.map(lambda x: x[0])
melt_data['Lat'] = GPS.map(lambda x: x[1])

melt_data.head(), melt_data.shape
```

```python
yearly = []  
for i in sorted(list(set(melt_data.Year_of_Release))):
    df_sub = melt_data[melt_data.Year_of_Release == i]
    df_sub = df_sub[df_sub.GeoLocation == 'Sales_NA']
    city = dict(
        type = 'scattergeo',
        locationmode = 'USA-states',
        visible=False,
        lon = df_sub['Lon'],
        lat = df_sub['Lat'],
        text = df_sub['GeoLocation'],
        marker = dict(
            size = df_sub['Sales'] * 10,
            color = ['rgb(215,25,28)', 'rgb(253,174,97)', 'rgb(26,150,65)', 'rgb(44,123,182)'],
            line = dict(width=0.5, color='rgb(40,40,40)'),
            sizemode = 'area'
        ),
        name = 'Year {}'.format(i) )
    yearly.append(city)

yearly = yearly[:-3]
yearly[5]['visible'] = True


steps1 = []
for i in range(len(yearly)):
    step = dict(
        method = 'restyle',
        label = 'lab1',
        args = ['visible', [False] * len(yearly)],
    )
    step['args'][1][i] = True # Toggle i'th trace to "visible"
    steps1.append(step)

    
steps2 = []
for i in sorted(list(set(melt_data.GeoLocation))):
    df_sub = melt_data[melt_data.GeoLocation == i]
    step = dict(
        method = 'update',
        label = 'lab2' + str(i),
        args = [{'marker': dict(
            size = df_sub['Sales'] * 10,
            color = ['rgb(215,25,28)'],
            line = dict(width=0.5, color='rgb(40,40,40)'),
            sizemode = 'area'), 'lon': df_sub['Lon'],
        'lat': df_sub['Lat'],
        'text': 'haha',}
                 ],
    )
    steps2.append(step)
    
sliders = []
sliders.append(dict(active = 5,
         currentvalue = {"prefix": "Year since 1980: "},
         pad = {"t": 0},
         steps = steps1))
sliders.append(dict(active = 2,
         currentvalue = {"prefix": "Year since 1999: "},
         pad = {"t": 100},
         steps = steps2))

layout = dict(sliders=sliders,
              title='Total Sales by year in each Region',
              xaxis=dict(title='Year'))

fig = dict(data=yearly, layout=layout)

py.iplot(fig, filename='Sine Wave Slider')
```
